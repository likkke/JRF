```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Figure7-author institution analysis.xlsx')
```

```python
k = list(df.loc[df['label']=="Corresponding author's institution vs. Highest ranked institution"].get('k'))

ORMS2018_first = list(df.loc[df['label']=="Corresponding author's institution vs. First author's institution"].get('ORMS2018'))
Economics2018_first = list(df.loc[df['label']=="Corresponding author's institution vs. First author's institution"].get('Economics2018'))
Mathematics2018_first = list(df.loc[df['label']=="Corresponding author's institution vs. First author's institution"].get('Mathematics2018'))
Physiology2018_first = list(df.loc[df['label']=="Corresponding author's institution vs. First author's institution"].get('Physiology2018'))

ORMS2018_highest = list(df.loc[df['label']=="Corresponding author's institution vs. Highest ranked institution"].get('ORMS2018'))
Economics2018_highest = list(df.loc[df['label']=="Corresponding author's institution vs. Highest ranked institution"].get('Economics2018'))
Mathematics2018_highest = list(df.loc[df['label']=="Corresponding author's institution vs. Highest ranked institution"].get('Mathematics2018'))
Physiology2018_highest = list(df.loc[df['label']=="Corresponding author's institution vs. Highest ranked institution"].get('Physiology2018'))
```

```python
fig,axes=plt.subplots(1,2, figsize=(6.85,3.42),dpi=600)

ax1=axes[0]
ax2=axes[1]

ax1.plot(k,ORMS2018_first,color='red',label='OR&MS, 2018',marker='s',lw=1,markersize=5)
ax1.plot(k,Economics2018_first,color='seagreen',label='Economics, 2018',marker='^',lw=1,markersize=5)
ax1.plot(k,Mathematics2018_first,color='dodgerblue',label='Mathematics, 2018',marker='o',lw=1,markersize=5)
ax1.plot(k,Physiology2018_first,color='orange',label='Physiology, 2018',marker='D',lw=1,markersize=5)


ax1.set_ylim(0.5,1)
ax1.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax1.set_ylabel("Correlatoin (Corresponding institution \nvs. First institution)", fontdict={'family' : 'ARIAL'})


ax2.plot(k,ORMS2018_highest,color='red',label='OR&MS, 2018',marker='s',lw=1,markersize=5)
ax2.plot(k,Economics2018_highest,color='seagreen',label='Economics, 2018',marker='^',lw=1,markersize=5)
ax2.plot(k,Mathematics2018_highest,color='dodgerblue',label='Mathematics, 2018',marker='o',lw=1,markersize=5)
ax2.plot(k,Physiology2018_highest,color='orange',label='Physiology, 2018',marker='D',lw=1,markersize=5)


ax2.set_ylim(0.5,1)
ax2.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax2.set_ylabel("Correlatoin (Corresponding institution \nvs. Highest ranked institution)", fontdict={'family' : 'ARIAL'})

ax1.text(-0.19, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.19, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)

ax1.legend(prop = {'family' : 'ARIAL'},ncol=2, bbox_to_anchor=[1.85,1.32])
plt.subplots_adjust(top =0.8,bottom=0.17, wspace=0.32)
plt.savefig('Figure7-author institution analysis.pdf') 
```
