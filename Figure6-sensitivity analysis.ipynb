```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Figure6-sensitivity analysis.xlsx')
```

```python
k= list(range(50,501,50))
data1 = df.loc[df['label']=='ORMS2018']
data2 = df.loc[df['label']=='Economics2018']
data3 = df.loc[df['label']=='Mathematics2018']
data4 = df.loc[df['label']=='Physiology2018']
```

```python
fig,axes=plt.subplots(2,2, figsize=(6.85,6.85),dpi=600)

ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]

colors = ['#FF6347', '#2E8B57', '#1E90FF']
sns.lineplot(x="k", y="y", hue="delta", data=data1,ci=95, palette=colors, ax = ax1)
ylabel ='Correlation (OR&MS, 2018)'
ax1.set_ylabel(ylabel,fontdict={'family' : 'ARIAL'})
ax1.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax1.legend_.remove()
ax1.set_xlim(25,525)
ax1.set_ylim(0.995,1.00005)


sns.lineplot(x="k", y="y", hue="delta", data=data2,ci=95, palette=colors, ax = ax2)
ylabel ='Correlation (Economics, 2018)'
ax2.set_ylabel(ylabel,fontdict={'family' : 'ARIAL'})
ax2.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax2.legend_.remove()
ax2.set_xlim(25,525)
ax2.set_ylim(0.995,1.00005)


sns.lineplot(x="k", y="y", hue="delta", data=data3,ci=95, palette=colors, ax = ax3)
ylabel ='Correlation (Mathematics, 2018)'
ax3.set_ylabel(ylabel,fontdict={'family' : 'ARIAL'})
ax3.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax3.legend_.remove()
ax3.set_xlim(25,525)
ax3.set_ylim(0.995,1.00005)


sns.lineplot(x="k", y="y", hue="delta", data=data4,ci=95, palette=colors, ax = ax4)
ylabel ='Correlation (Physiology, 2018)'
ax4.set_ylabel(ylabel,fontdict={'family' : 'ARIAL'})
ax4.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax4.legend_.remove()
ax4.set_xlim(25,525)
ax4.set_ylim(0.995,1.00005)

handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles=handles[1:], labels=['$\Delta = 10\%$', '$\Delta = 30\%$', '$\Delta = 50\%$'],prop = {'family' : 'ARIAL'},ncol=3, bbox_to_anchor=[2.1,1.23])

ax1.text(-0.30, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.30, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)
ax3.text(-0.30, 1.02, 'c',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax3.transAxes)
ax4.text(-0.30, 1.02, 'd',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax4.transAxes)  
plt.subplots_adjust(hspace =0.25,wspace=0.38)

plt.savefig('Figure6-sensitivity analysis plot.pdf') 
```
