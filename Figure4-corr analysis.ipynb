```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Figure4-corr data.xlsx')
```

```python
k = list(df.loc[df['label']=='JRF vs. AIS'].get('k'))
ORMS2014_IF = list(df.loc[df['label']=='JRF vs. IF'].get('ORMS2014'))
ORMS2015_IF = list(df.loc[df['label']=='JRF vs. IF'].get('ORMS2015'))
ORMS2016_IF = list(df.loc[df['label']=='JRF vs. IF'].get('ORMS2016'))
ORMS2017_IF = list(df.loc[df['label']=='JRF vs. IF'].get('ORMS2017'))
ORMS2018_IF = list(df.loc[df['label']=='JRF vs. IF'].get('ORMS2018'))
Economics2018_IF = list(df.loc[df['label']=='JRF vs. IF'].get('Economics2018'))
Mathematics2018_IF = list(df.loc[df['label']=='JRF vs. IF'].get('Mathematics2018'))
Physiology2018_IF = list(df.loc[df['label']=='JRF vs. IF'].get('Physiology2018'))

ORMS2014_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('ORMS2014'))
ORMS2015_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('ORMS2015'))
ORMS2016_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('ORMS2016'))
ORMS2017_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('ORMS2017'))
ORMS2018_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('ORMS2018'))
Economics2018_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('Economics2018'))
Mathematics2018_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('Mathematics2018'))
Physiology2018_AIS = list(df.loc[df['label']=='JRF vs. AIS'].get('Physiology2018'))
```

```python
fig,axes=plt.subplots(2,2, figsize=(6.85,6.85),dpi=600)

ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]

ax1.plot(k,ORMS2018_IF,color='red',label='OR&MS, 2018',marker='s',lw=1,markersize=5)
ax1.plot(k,Economics2018_IF,color='seagreen',label='Economics, 2018',marker='^',lw=1,markersize=5)
ax1.plot(k,Mathematics2018_IF,color='dodgerblue',label='Mathematics, 2018',marker='o',lw=1,markersize=5)
ax1.plot(k,Physiology2018_IF,color='orange',label='Physiology, 2018',marker='D',lw=1,markersize=5)
ax1.legend(prop={'family' : 'ARIAL'},loc=4)
ax1.set_xticks([200,400,600,800,1000])
ax1.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax1.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax1.set_ylabel('Correlatoin (JRF vs. 5-year IF)', fontdict={'family' : 'ARIAL'})


ax2.plot(k,ORMS2018_AIS,color='red',label='OR&MS, 2018',marker='s',lw=1,markersize=5)
ax2.plot(k,Economics2018_AIS,color='seagreen',label='Economics, 2018',marker='^',lw=1,markersize=5)
ax2.plot(k,Mathematics2018_AIS,color='dodgerblue',label='Mathematics, 2018',marker='o',lw=1,markersize=5)
ax2.plot(k,Physiology2018_AIS,color='orange',label='Physiology, 2018',marker='D',lw=1,markersize=5)
ax2.legend(prop={'family' : 'ARIAL'},loc=4)
ax2.set_xticks([200,400,600,800,1000])
ax2.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax2.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax2.set_ylabel('Correlatoin (JRF vs. 5-year AIS)', fontdict={'family' : 'ARIAL'})


ax3.plot(k,ORMS2014_IF,color='seagreen',label='OR&MS, 2014',marker='^',lw=1,markersize=5)
ax3.plot(k,ORMS2015_IF,color='dodgerblue',label='OR&MS, 2015',marker='d',lw=1,markersize=5)
ax3.plot(k,ORMS2016_IF,color='orange',label='OR&MS, 2016',marker='o',lw=1,markersize=5)
ax3.plot(k,ORMS2017_IF,color='blueviolet',label='OR&MS, 2016',marker='D',lw=1,markersize=5)
ax3.plot(k,ORMS2018_IF,color='red',label='OR&MS, 2017',marker='s',lw=1,markersize=5)
ax3.legend(prop={'family' : 'ARIAL'},loc=4)
ax3.set_xticks([200,400,600,800,1000])
ax3.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax3.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax3.set_ylabel('Correlatoin (JRF vs. 5-year IF)', fontdict={'family' : 'ARIAL'})

ax4.plot(k,ORMS2014_AIS,color='seagreen',label='OR&MS, 2014',marker='^',lw=1,markersize=5)
ax4.plot(k,ORMS2015_AIS,color='dodgerblue',label='OR&MS, 2015',marker='d',lw=1,markersize=5)
ax4.plot(k,ORMS2016_AIS,color='orange',label='OR&MS, 2016',marker='o',lw=1,markersize=5)
ax4.plot(k,ORMS2017_AIS,color='blueviolet',label='OR&MS, 2016',marker='D',lw=1,markersize=5)
ax4.plot(k,ORMS2018_AIS,color='red',label='OR&MS, 2017',marker='s',lw=1,markersize=5)
ax4.legend(prop={'family' : 'ARIAL'},loc=4)
ax4.set_xticks([200,400,600,800,1000])
ax4.set_yticks([0,0.2,0.4,0.6,0.8,1])
ax4.set_xlabel('k',fontdict={'family' : 'ARIAL'})
ax4.set_ylabel('Correlatoin (JRF vs. 5-year AIS)', fontdict={'family' : 'ARIAL'})

ax1.text(-0.18, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.18, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)
ax3.text(-0.18, 1.02, 'c',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax3.transAxes)
ax4.text(-0.18, 1.02, 'd',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax4.transAxes)

plt.subplots_adjust(wspace=0.25)
plt.savefig('Figure4-corr plot.pdf') 
```
