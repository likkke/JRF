```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel(io = r'Figure2-IF_case.xlsx') 
df2 = pd.read_excel(io = r'Figure2-JRF_case.xlsx') 
```

```python
x = [i for i in range(0,1001)]
ax1_omega = df1.OMEGA
ax2_omega = df2.OMEGA

ax1_ms = df1.MS
ax2_ms = df2.MS

ax1_or = df1.OR
ax2_or = df2.OR
```

```python
fig,axes=plt.subplots(1,2, figsize=(6.85,3.42),dpi=600)
colors = ['tomato', 'seagreen', 'dodgerblue']
ax1=axes[0]
ax2=axes[1]

ax1.plot(x,ax1_omega,c = colors[0], label = 'Omega',ls='-')
ax1.plot(x,ax1_ms,c = colors[1], label = 'MS',ls='--')
ax1.plot(x,ax1_or,c = colors[2], label = 'OR',ls=':')
ax1.set_xlabel('k',fontdict = {'family' : 'ARIAL'})
ax1.set_ylabel('$\mathregular{IF(k)^5_{2018}}$',fontdict = {'family' : 'ARIAL'})
ax1.set_xlim(0,1000)
ax1.set_ylim(0,4)
ax1.legend(loc = 0, prop = {'family' : 'ARIAL'})

ax2.plot(x,ax2_omega,c = colors[0], label = 'Omega',ls='-')
ax2.plot(x,ax2_ms,c = colors[1], label = 'MS',ls='--')
ax2.plot(x,ax2_or,c = colors[2], label = 'OR',ls=':')
ax2.set_xlim(0,1000)
ax2.set_ylim(0,3)
ax2.set_xlabel('k',fontdict = {'family' : 'ARIAL'})
ax2.set_ylabel('$\mathregular{JRF(k)^5_{2018}}$',fontdict = {'family' : 'ARIAL'})
ax2.legend(loc = 0, prop = {'family' : 'ARIAL'})

ax1.text(-0.19, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.19, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)

plt.subplots_adjust(wspace=0.29)
plt.savefig('Figure2-IF and JRF of MS OR and Ome.pdf') 
```
