```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Figure3-paper and citation institution distribution.xlsx')
```

```python
k = list(df.loc[df['label']=='paper'].get('k'))

Omega_paper = list(df.loc[df['label']=='paper'].get('Omega'))
MS_paper = list(df.loc[df['label']=='paper'].get('MS'))
OR_paper = list(df.loc[df['label']=='paper'].get('OR'))

Omega_citation = list(df.loc[df['label']=='citation'].get('Omega'))
MS_citation = list(df.loc[df['label']=='citation'].get('MS'))
OR_citation = list(df.loc[df['label']=='citation'].get('OR'))
```

```python
fig,ax=plt.subplots(2,1, figsize=(6.85,6.85),dpi=600)
ax1=ax[0]
ax2=ax[1]

width = 12.5
k1 = [i-width for i in k]
k2 = k
k3 = [i+width for i in k]

colors = ['tomato', 'seagreen', 'dodgerblue']
ax1.bar(k1, Omega_paper, width=width, label='Omega',fc = colors[0])
ax1.bar(k2, MS_paper, width=width, label='MS',fc = colors[1])
ax1.bar(k3, OR_paper, width=width, label='OR',fc = colors[2])
name= [str(int(i)-49)+'-'+str(i) for i in k]
ax1.set_xticks(k)
ax1.set_xticklabels(name,rotation=60)
ax1.set_ylabel('CPP',fontdict={'family' : 'ARIAL'})
ax1.legend(prop = {'family' : 'ARIAL'},ncol=3, bbox_to_anchor=[0.8,1.17])

ax2.bar(k1, Omega_citation, width=width, label='Omega',fc = colors[0])
ax2.bar(k2, MS_citation, width=width, label='MS',fc = colors[1])
ax2.bar(k3, OR_citation, width=width, label='OR',fc = colors[2])
name= [str(int(i)-49)+'-'+str(i) for i in k]
ax2.set_xticks(k)
ax2.set_xticklabels(name,rotation=60)
ax2.set_ylabel('The percentage of publications',fontdict={'family' : 'ARIAL'})

ax1.text(-0.05, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.05, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)



plt.subplots_adjust(hspace=0.35)
plt.savefig('Figure3-paper and citation institution distribution.pdf') 
```
