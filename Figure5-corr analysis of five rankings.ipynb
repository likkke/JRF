```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Figure5-corr data based on five rankings.xlsx')
```

```python
k = list(df.loc[df['label']=='ORMS2018'].get('k'))

createVar = locals()
labels = ['ORMS2018', 'Economics2018', 'Mathematics2018', 'Physiology2018']
for label in labels:
    createVar['ARWUvsQS_{}'.format(label)] = list(df.loc[df['label']==label].get('ARWU vs. QS'))
    createVar['ARWUvsTIMES_{}'.format(label)] = list(df.loc[df['label']==label].get('ARWU vs. TIMES'))
    createVar['ARWUvsUS_{}'.format(label)] = list(df.loc[df['label']==label].get('ARWU vs. US'))
    createVar['ARWUvsPR_{}'.format(label)] = list(df.loc[df['label']==label].get('ARWU vs. PR'))
```

```python
fig,axes=plt.subplots(2,2, figsize=(6.85,6.85),dpi=100)

ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]

colors = ['tomato', 'seagreen', 'dodgerblue','orange']
for num, ax in enumerate([ax1, ax2, ax3, ax4]):
    ax.plot(k, createVar['ARWUvsQS_{}'.format(labels[num])], marker= '^', color=colors[0],lw=1,markersize=5,label='ARWU vs. QS ranking')
    ax.plot(k, createVar['ARWUvsTIMES_{}'.format(labels[num])], marker= 'd', color=colors[1],lw=1,markersize=5,label='ARWU vs. THE ranking')
    ax.plot(k, createVar['ARWUvsUS_{}'.format(labels[num])], marker= 'o', color=colors[2],lw=1,markersize=5,label='ARWU vs. U.S. News ranking')
    ax.plot(k, createVar['ARWUvsPR_{}'.format(labels[num])], marker= 'D', color=colors[3],lw=1,markersize=5,label='ARWU vs. EF ranking')
    if labels[num]== 'ORMS2018':
        ylabel ='Correlation (OR&MS, 2018)'
    else:
        ylabel ='Correlation ('+labels[num][:-4]+', 2018)'
    ax.set_ylabel(ylabel,fontdict={'family' : 'ARIAL'})
    ax.set_xlabel('k',fontdict={'family' : 'ARIAL'})
    ax.set_xlim(25,525)
    ax.set_ylim(0.85,1.003)
ax1.legend(prop = {'family' : 'ARIAL'},ncol=2, bbox_to_anchor=[2.1,1.3])

ax1.text(-0.21, 1.02, 'a',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax1.transAxes)
ax2.text(-0.21, 1.02, 'b',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax2.transAxes)
ax3.text(-0.21, 1.02, 'c',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax3.transAxes)
ax4.text(-0.21, 1.02, 'd',horizontalalignment='center',verticalalignment='center',fontdict={'family' : 'ARIAL','size': 15},transform=ax4.transAxes)  

plt.subplots_adjust(hspace =0.25,wspace=0.28)
plt.savefig('Figure5-corr analysis of five rankings.pdf') 
```
